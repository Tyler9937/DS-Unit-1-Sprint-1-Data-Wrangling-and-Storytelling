```python
#importing essential libraries
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns

#created model dataframe for tidy data testing
table1 = pd.DataFrame(
    [[np.nan, 2],
     [16,    11], 
     [3,      1]],
    index=['John Smith', 'Jane Doe', 'Mary Johnson'], 
    columns=['treatmenta', 'treatmentb'])

table2 = table1.T
```

```python
# Take the row index, and add it as a new column
table1 = table1.reset_index()
table1
```

```python
# using .melt() to 'combine both treatment into one variable
tidy1 = table1.melt(id_vars='index', value_vars=['treatmenta', 'treatmentb'])
tidy1
```

```python
#renaming column names
tidy1 = tidy1.rename(columns={
    'index': 'name', 
    'variable': 'trt', 
    'value': 'result'
})
tidy1
```

```python
#using .replace() to simplify data
tidy1.trt = tidy1.trt.str.replace('treatment', '')

tidy1
```

```python
##### YOUR CODE HERE #####
table2 = table2.reset_index()
table2
```

```python
#using .melt() to combine names
tidy2 = table2.melt(id_vars='index', value_vars=['John Smith', 'Jane Doe', 'Mary Johnson'])
tidy2
```

```python
#renaming columns
tidy2 = tidy2.rename(columns={
    'index': 'trt',
    'variable': 'name',
    'value': 'result'
})
tidy2
```

```python
#Replacig treatment to tidy up
tidy2.trt = tidy2.trt.str.replace('treatment', '')
tidy2
```

```python
tidy1
```

```python
#using .pivot_table to 'inverse data
wide = tidy1.pivot_table(index='name', columns='trt', values='result')
wide
```

```python
#using .pivot_table to inverse data
wide2 = tidy2.pivot_table(index='trt', columns='name', values='result')
wide2
```
